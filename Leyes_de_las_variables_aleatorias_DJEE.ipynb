```python
#Librerias
from itertools import product
from fractions import Fraction
from collections import defaultdict
import matplotlib.pyplot as plt
```

```python
#### SUMA ####

#Función Probabilidad global
def P(A, Omega):
  P= Fraction(len(A),len(Omega))
  return P

## Crear espacio muestral Omega
L= [i for i in range(1,7)] #Lista por comprensión

Omega= set(product(L, repeat=2))

S= {(i,j): i+j for i,j in Omega} #suma

```

```python
# Crear diccionario por default de manera sintetica y arreglada
from collections import defaultdict
dS= defaultdict(set) #lista

for i,j in S.items():
  dS[j].add(i) #añadir al diccionario las llaves

leyS= {k:P(A,Omega) for k,A in dS.items()}

sum_probS = sum(leyS.values())
verificacionS = sum_probS == 1

print("Omega =", Omega, "\n", 'Cardinalidad =', len(Omega))

print('--------------------------------------------------------',"\n")
print(leyS)
```

```python
print("Suma de las probabilidades =", sum_probS)
print("¿La suma es igual a 1? =", verificacionS)
```

### 1 GRAFICA

```python
suma = list(leyS.keys())
probabilidades = [float(leyS[k]) for k in suma]

# Gráfico
plt.figure(figsize=(10, 6))
plt.bar(suma, probabilidades, color='green', edgecolor='black')

plt.title('Distribución de Probabilidades de la Suma de Dos Dados')
plt.xlabel('Suma de las Caras')
plt.ylabel('Probabilidad')
plt.xticks(suma)  # Mostrar cada valor en el eje x

# Imprime gráfica
plt.grid(True)
plt.show()
```

## La diferencia de las caras es mayor que 1

```python

def P(A, OmegaD):
     P=Fraction(len(A), len(OmegaD))
     return P
# Espacio muestral Omega
LD = [i for i in range(1, 7)]
OmegaD = set(product(LD, repeat=2))
```

```python
#PRIMERO LAS DIFERENCIAS
# Crear diccionario con la diferencias
D = {(i, j): abs(i - j) for i, j in OmegaD}

# diccionario por default
difD = defaultdict(set)

for (i, j), diff in D.items():
    difD[diff].add((i, j))

# Calcular la ley de probabilidad para cada diferencia posible
leyD = {k: P(A, OmegaD) for k, A in difD.items()}

# Verificar que la suma de las probabilidades sea igual a 1
sum_prob_diferencia = sum(leyD.values())
verificaciond = sum_prob_diferencia == 1

print("Omega =", OmegaD, "\n", 'Cardinalidad =', len(OmegaD))
```

```python
print("Suma de las probabilidades =", sum_prob_diferencia)
print("¿La suma es igual a 1? =", verificaciond)
```

```python
#Ahora filtramos las diferencias que sea mayores a 1
dD_filtro = {k: A for k, A in difD.items() if k > 1}

# Calcular la  probabilidad para las diferencias >1
leyD_filtro = {k: P(A, OmegaD) for k, A in dD_filtro.items()}

# la suma de las probabilidades filtradas =  1
sum_p_filtro = sum(leyD_filtro.values())
ver_filtro = sum_p_filtro == 1

print("Omega =", leyD_filtro, "\n", 'Cardinalidad =', len(leyD_filtro))
```

```python
print("Suma de las probabilidades >1 =", sum_p_filtro)
print("¿La suma = 1 para diferencias > 1? =", ver_filtro)
```

### 1.1 Gráfica

```python
sumaf = list(leyD_filtro.keys())
p_f = [float(leyD_filtro[k]) for k in sumaf]  # Convertir a flotante para la gráfica

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.bar(sumaf, p_f, color='green', edgecolor='black')

# Añadir títulos y etiquetas
plt.title('Distribución de Probabilidades de diferencias de caras >1')
plt.xlabel('Diferencia de las Caras')
plt.ylabel('Probabilidad')
plt.xticks(sumaf)  # Mostrar cada valor de suma en el eje x

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

### **La suma de caras sea par**

```python
def P(A, Omega):
    return Fraction(len(A), len(Omega))

# Crear Omega
Lp = [i for i in range(1, 7)]
Omegapar = set(product(Lp, repeat=2))
print("Omega suma par =", Omegapar, "\n", 'Cardinalidad =', len(Omegapar))
```

```python
# suma de las caras de los dados
Sp = {(i, j): i + j for i, j in Omegapar}

# Crear diccionario por default para almacenar las combinaciones según la suma de las caras
dSp = defaultdict(set)

for (i, j), suma in Sp.items():
    dSp[suma].add((i, j))

# Filtrar suma de las caras es par
dS_par = {k: A for k, A in dSp.items() if k % 2 == 0}

# Calcular la probabilidad para par
leyS_par = {k: P(A, Omega) for k, A in dS_par.items()}

# Verificar la suma
sumaprob_par = sum(leyS_par.values())
ver_par = sumaprob_par == 1

print("\nProbabilidades de sumas pares:", leyS_par)
print("Suma de las probabilidades (pares) =", sumaprob_par)
print("¿La suma es igual a 1 para sumas pares? =", ver_par)
```

```python
## ACUMULARIVA =1

# Función de distribución acumulativa (CDF) de la suma de los resultados
# Verificar espacio muestral completo
leySp = {k: P(A, Omega) for k, A in dSp.items()}
sum_prob_c = sum(leySp.values())
ver_c = sum_prob_c == 1
print("\nProbabilidad completa:", leySp)
print("Suma de las probabilidades =", sum_prob_c)
print("¿La suma es igual a 1 para el espacio muestral completo? =", ver_c)
```

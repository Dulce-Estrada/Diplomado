```python
#Librerias
from itertools import product
from fractions import Fraction
from collections import defaultdict
import matplotlib.pyplot as plt
```

```python
#### SUMA ####

#Función Probabilidad global
def P(A, Omega):
  P= Fraction(len(A),len(Omega))
  return P

## Crear espacio muestral Omega
L= [i for i in range(1,7)] #Lista por comprensión

Omega= set(product(L, repeat=2))

S= {(i,j): i+j for i,j in Omega} #suma

```

```python
# Crear diccionario por default de manera sintetica y arreglada
from collections import defaultdict
dS= defaultdict(set) #lista

for i,j in S.items():
  dS[j].add(i) #añadir al diccionario las llaves

leyS= {k:P(A,Omega) for k,A in dS.items()}

sum_probS = sum(leyS.values())
verificacionS = sum_probS == 1

print("Omega =", Omega, "\n", 'Cardinalidad =', len(Omega))

print('--------------------------------------------------------',"\n")
print(leyS)
```

```python
print("Suma de las probabilidades =", sum_probS)
print("¿La suma es igual a 1? =", verificacionS)
```

```python
suma = list(leyS.keys())
probabilidades = [float(leyS[k]) for k in suma]

# Gráfico
plt.figure(figsize=(10, 6))
plt.bar(suma, probabilidades, color='green', edgecolor='black')

plt.title('Distribución de Probabilidades de la Suma de Dos Dados')
plt.xlabel('Suma de las Caras')
plt.ylabel('Probabilidad')
plt.xticks(suma)  # Mostrar cada valor en el eje x

# Imprime gráfica
plt.grid(True)
plt.show()
```

## La diferencia de las caras es mayor que 1

```python

def P(A, OmegaD):
     P=Fraction(len(A), len(OmegaD))
     return P
# Espacio muestral Omega
LD = [i for i in range(1, 7)]
OmegaD = set(product(LD, repeat=2))
```

```python
#PRIMERO CREAMOS LAS DIFERENCIAS
# Crear diccionario con la diferencias
D = {(i, j): abs(i - j) for i, j in OmegaD}

# diccionario por default
difD = defaultdict(set)

for (i, j), diff in D.items():
    difD[diff].add((i, j))

# Calcular la ley de probabilidad para cada diferencia posible
leyD = {k: P(A, OmegaD) for k, A in difD.items()}

# Verificar que la suma de las probabilidades sea igual a 1
sum_prob_diferencia = sum(leyD.values())
verificaciond = sum_prob_diferencia == 1

print("Omega =", OmegaD, "\n", 'Cardinalidad =', len(OmegaD))
```

```python
print("Suma de las probabilidades =", sum_prob_diferencia)
print("¿La suma es igual a 1? =", verificaciond)
```

```python
#Ahora filtramos las diferencias que sea mayores a 1
dD_filtro = {k: A for k, A in difD.items() if k > 1}

# Calcular la ley de probabilidad para las diferencias filtradas
leyD_filtro = {k: P(A, OmegaD) for k, A in dD_filtro.items()}

# Verificar que la suma de las probabilidades filtradas sea igual a 1
sum_p_filtro = sum(leyD_filtro.values())
ver_filtro = sum_p_filtro == 1

print("Omega =", leyD_filtro, "\n", 'Cardinalidad =', len(leyD_filtro))
```

```python
print("Suma de las probabilidades filtradas =", sum_p_filtro)
print("¿La suma es igual a 1 para diferencias mayores que 1? =", ver_filtro)
```

### Demostraciones de una Distribución Uniforme Discreta

Sea una variable aleatoria discreta $X$ que sigue una distribución uniforme discreta en el conjunto $({X_1, X_1 + 1, \ldots, X_n})$. La función de masa de probabilidad (PMF) es:

$$ P(X = k) = \frac{1}{X_n - X_1 + 1} $$

#### **1. Media**

La media $E(X)$ se calcula como:

$$ E(X) = \sum_{k=X_1}^{X_n} k \cdot P(X = k) $$

Sustituyendo la PMF:

$$ E(X) = \sum_{k=X_1}^{X_n} k \cdot \frac{1}{X_n - X_1 + 1} $$

Como $( \frac{1}{X_n - X_1 + 1} )$ es una constante, se puede sacar fuera de la suma:

$$ E(X) = \frac{1}{X_n - X_1 + 1} \sum_{k=X_1}^{X_n} k $$

La suma de una secuencia aritmética es:

$$ \sum_{k=X_1}^{X_n} k = \frac{(X_n - X_1 + 1)(X_1 + X_n)}{2} $$

Por lo tanto:

$$ E(X) = \frac{1}{X_n - X_1 + 1} \cdot \frac{(X_n - X_1 + 1)(X_1 + X_n)}{2} $$

Simplificando:

$$ E(X) = \frac{X_1 + X_n}{2} $$

#### **2. Varianza**

La varianza ${Var}(X)$ se calcula como:

$$ \text{Var}(X) = E(X^2) - \left(E(X)\right)^2 $$

Primero, calculamos $E(X^2)$:

$$ E(X^2) = \sum_{k=X_1}^{X_n} k^2 \cdot P(X = k) $$

Sustituyendo la PMF:

$$ E(X^2) = \frac{1}{X_n - X_1 + 1} \sum_{k=X_1}^{X_n} k^2 $$

La suma de los cuadrados es:

$$ \sum_{k=X_1}^{X_n} k^2 = \frac{(X_n - X_1 + 1)(X_1^2 + X_n^2 + X_1 X_n)}{3} $$

Entonces:

$$ E(X^2) = \frac{1}{X_n - X_1 + 1} \cdot \frac{(X_n - X_1 + 1)(X_1^2 + X_n^2 + X_1 X_n)}{3} $$

$$ E(X^2) = \frac{X_1^2 + X_n^2 + X_1 X_n}{3} $$

La varianza es:

$$ {Var}(X) = E(X^2) - \left(E(X)\right)^2 $$


Sustituyendo $E(X) = \frac{X_1 + X_n}{2}$ y $E(X^2)$ :

$$ \text{Var}(X) = \frac{X_1^2 + X_n^2 + X_1 X_n}{3} - \left(\frac{X_1 + X_n}{2}\right)^2 $$

$\left(\frac{X_1 + X_n}{2}\right)^2$:
$ \left(\frac{X_1 + X_n}{2}\right)^2 = \frac{X_1^2 + 2X_1 X_n + X_n^2}{4} $

Entonces:

$$ \text{Var}(X) = \frac{X_1^2 + X_n^2 + X_1 X_n}{3} - \frac{X_1^2 + 2X_1 X_n + X_n^2}{4} $$

Simplificando:

$$ \text{Var}(X) = \frac{4(X_1^2 + X_n^2 + X_1 X_n) - 3(X_1^2 + 2X_1 X_n + X_n^2)}{12} $$

$$ \text{Var}(X) = \frac{(X_n - X_1 + 1)^2 - 1}{12} $$
